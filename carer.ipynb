```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('datajurusan.csv')
df
```

```python
df['minat_bakat'] = df['minat'] + ' ' + df['bakat']
df
```

```python
df_filter = df[['jurusan', 'minat_bakat']]
df_filter.head()
```

```python
df_minatbakat = df.minat_bakat
df_minatbakat
```

```python
def remove(data):
    data = str(data).lower()
    data = data.replace('[', '')
    data = data.replace(']', '')
    data = data.replace(',', '')
    data = data.replace("'", '')
    return data

df_minatbakat = df_minatbakat.apply(remove)
df_minatbakat
```

```python
df_minatbakat.to_csv("minatbat_filtered.csv")
df_filter.to_csv("data_filtered.csv")
```

```python
vector_tfidf = TfidfVectorizer()
```

```python
text = 'sosiologi, sejarah'
```

```python
df_minatbakat_new = pd.concat([df_minatbakat, pd.Series([text])], ignore_index=True)

result_vector = vector_tfidf.fit_transform(df_minatbakat_new)
result_array  = result_vector.toarray()
```

```python
result_array[:-1].shape
```

```python
result = {}

for id, vector in enumerate(result_array[:-1]):
    cosine_val = cosine_similarity([result_array[id]], [result_array[-1]])
    result[id] = cosine_val

result_desc = dict(sorted(result.items(), key=lambda item: item[1], reverse=True))

top_n = 5

for n, no_jurusan in enumerate(result_desc):
    if n > 5:
        break
    print(df_filter.iloc[no_jurusan]['jurusan'])
```
